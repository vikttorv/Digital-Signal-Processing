```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import matplotlib.ticker as ticker
def DTFT(x, M):
    """
    Функция вычисляет значения ДВПФ в M точках на отрезке 
    по оси нормированных частот [-0.5; 0.5]

    Parameters
    ----------
    x : complex
        входная последовательность отсчетов, первый отсчет при k=0
    M : int
        колличество точек на отрезке [-0.5; 0.5]

    Returns
    -------
    float
        значения оси нормированных частот, 
        соотвествующих вычисленным значениям
        спектральной функции  
    complex
        значения спектральной функции      
    """
    return (-0.5+(np.arange(M)/M), np.fft.fftshift(np.fft.fft(x, M)))
```

**Алгоритм Герцеля**

1) Для $m=0,\,1,\, \dots,\, N-1$ определяем 

$${{V}_{n}}[m]=x[m]+2\cos (2\pi n/N){{V}_{n}}[m-1]-{{V}_{n}}[m-2],\quad 0\le m\le N-1,$$

2) Вычисляем $X[n]$ (при $m=N-1$):

$$X[n]={{\left. {{y}_{n}}[m] \right|}_{m=N-1}}={{\left. \left( W_{N}^{-n}{{V}_{n}}[m]-{{V}_{n}}[m-1] \right) \right|}_{m=N-1}}.$$
$$V_n[-1]=V_n [-2] = 0$$


**Алгоритм Герцеля**

1) Для $m=0,\,1,\, \dots,\, N-1$ определяем 

$${{V}_{n}}[m]=x[m]+2\cos (2\pi n/N){{V}_{n}}[m-1]-{{V}_{n}}[m-2],\quad 0\le m\le N-1,$$

2) Вычисляем $X[n]$ (при $m=N-1$):

$$X[n]={{\left. {{y}_{n}}[m] \right|}_{m=N-1}}={{\left. \left( W_{N}^{-n}{{V}_{n}}[m]-{{V}_{n}}[m-1] \right) \right|}_{m=N-1}}.$$
$$V_n[-1]=V_n [-2] = 0$$

```python
def gertzel (sig, sample):
    N = sig.size
    assert N > 0
    assert sample >= 0 and sample < N

    V = np.array ([0 + 1j] * N)
    V[0] = sig[0]
    if N > 1:
        V[1] = sig[1] + 2 * np.cos (2 * np.pi * sample / N) * V[0]
    for k in range (2, N, 1):
        V[k] = sig[k] + 2 * np.cos (2 * np.pi * sample / N) * V[k - 1] - V[k - 2] 
    
    y = 0
    if N > 1:
        y = np.exp (2 * np.pi * 1j * sample / N) * V[N - 1] - V[N - 2]
    else:
        y = np.exp (2 * np.pi * 1j * sample / N) * V[N - 1]
    return y
```

```python
n1 = 7
n2 = 143
n3 = 1052

start_time = time.time ()
value1 = gertzel (x, n1)
print ("The time of computation of Gertzel algorithm for n = {} is {} seconds.".format (n1, time.time () - start_time))
print ("The difference from FFT is {}".format (abs(value1 - X2[n1])))

start_time = time.time ()
value2 = gertzel (x, n2)
print ("The time of computation of Gertzel algorithm for n = {} is {} seconds.".format (n2, time.time () - start_time))
print ("The difference from FFT is {}".format (abs(value2 - X2[n2])))

start_time = time.time ()
value3 = gertzel (x, n3)
print ("The time of computation of Gertzel algorithm for n = {} is {} seconds.".format (n3, time.time () - start_time))
print ("The difference from FFT is {}".format (abs(value3 - X2[n3])))
```

```python
num_per_segment = 2205
num_overlap = 1103
num_fft = 10000
f, t, Zxx = signal.stft(x, fs=fs, window=('hann'), nperseg=num_per_segment, noverlap=num_overlap, nfft=num_fft)
plt.figure(figsize=[13.5, 5], dpi=70)
plt.pcolormesh(t, f, np.abs (Zxx), cmap=plt.get_cmap('Purples'), shading='flat')
plt.colorbar()
plt.ylabel('Частота, Гц')
plt.xlabel('Время, с')
plt.ylim((0, 1500))
plt.tight_layout()
```

1) - Исходя из спектрограммы можно заключить, что сигнал состоит из двух гармонических компонент, которые повторяются 6 раз (6 звоночков) и в каждый из 6 звоночков эти гармонические компоненты постепенно затухают. При этом частоты гармонических компонент в каждом последующем звоночке меньше, чем в предыдущем.

```python
window_duration = num_per_segment / fs
overlap_duration = num_overlap / fs
print ("The duration of segment is {} s".format (window_duration))
print ("The duration of of overlap is {} s".format (overlap_duration))
```

2) - Гранулярность по оси X определяется длиной сегмента без overlap-а, там где должен быть следующий сегмент ничего от старого уже не отрисовывается. Хорошо видно, что график сегментирован по оси X. Я специально подобрал "хорошие" размеры сегмента и перекрытия, чтобы показать, что в интервале $[0, 0.25]$ лежит 10, а не 5 кусочков.

3) - Разрешение по частоте равно $f_s / N_{fft}$, а по времени равно $(N_{perseg} - N_{overlap}) / f_s$ секунд.

```python
print ("Time resolution is {} Hz".format (fs / num_fft))
print ("Frequency resolution is {} s".format ((num_per_segment - num_overlap) / fs))
```

Постройте периодограмму для оценки СПМ процесса, по выборкам, которые могут быть описаны формулой

где $N=64$, $\varepsilon[k]$ — дискретный белый гауссовский шум c нулевым средним и с дисперсией $\sigma^2=400$ (`scale=20`), $a_1=5$, частота дискретизации  $f_s=16$ кГц, $f_0=2$ кГц.

Выделение гармонической компоненты на фоне шума невозможно. 

Оцениваем СПМ случайного процесса
$${{x}_{\text{сл}}}[k]={A}\sin (2\pi {{f}_{1}}k\Delta t+{{\phi }_{1}})+e[k]$$
где 
* $A$ — известная амплитуда,
* $e[k]$ — дискретный белый шум c нулевым средним и дисперсией ${{\sigma }^{2}}$, 
* ${{f}_{1}}=2\ \text{кГц}$,  ${{f}_{\text{д}}}=1/\Delta t=10\ \text{Гц}$, 
* ${\phi }_{1}$ — случайная величина с равномерным распределением на отрезке $[0,2\pi ]$.

На частоте $f_1$ по формуле из лекции:

$M\left\{ \hat{G}\left( f_1 \right) \right\}=\Delta t\dfrac{{{A}^{2}}}{4U} |W(0)|^2 +\sigma^2 \ \Delta t$

$U=\int\limits_{-1/2}^{\text{1}/2}{{{\left| W(\nu ) \right|}^{2}}}d\nu =\sum\limits_{k=0}^{N-1}{{{w}^{2}}[k]}$

Данные формулы можно легко получить в уме, если знать выражение для модифицированной периодограммы Шустера:

$$\hat{G}(f)=\frac{\Delta t}{U}{{\left| \sum\limits_{k\ =\ 0}^{N-1}{x[k]w[k]\,{{e}^{-j2\pi fk\Delta t}}} \right|}^{2}}.$$ 	

Для прямоугольного окна $W(0)=N=U$, где $N$ — длина окна. Тогда на частоте $f_1$

$M\left\{ \hat{G}\left( f_1 \right) \right\}=\Delta t  \left( \dfrac{{{A}^{2}}}{4}N+\sigma^2 \right)$

В итоге искомое N равно:

$N = \dfrac{40 \cdot 4\sigma^2}{A^2} = 2560$

```python
fs=16000
f1=2000
A=5
N=2560 
k=np.arange(N)
x=A*np.sin(2*np.pi*f1*k/fs)+np.random.normal(scale=20, size=N)
```
